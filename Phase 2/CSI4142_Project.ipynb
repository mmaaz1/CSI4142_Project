#**Date Dimension**

**Create date dimension**

```python
start_date = '2019-01-01'
end_date = '2022-12-31'
dates = pd.date_range(start=start_date, end=end_date, freq='D')
date_dim = pd.DataFrame({'Date': dates})
date_dim['Year'] = date_dim['Date'].dt.year
date_dim['Month'] = date_dim['Date'].dt.month
date_dim['Week'] = date_dim['Date'].dt.isocalendar().week
date_dim['Day'] = date_dim['Date'].dt.day
date_dim['Date'] = date_dim['Date'].astype('string')
```

**Remove non-alphanumeric values from Country (For Merging)**

**Only taking the necessary columns**

```python
env_df = env_df.loc[:, ['year', 'month', 'average_CO2', 'average_CO4', 'average_N2O']]
```

**Check Type**

```python
env_df.dtypes
```

**Drop Rows with Null facts**

```python
env_df.dropna(subset=['average_CO2'], inplace=True)
env_df.dropna(subset=['average_CO4'], inplace=True)
env_df.dropna(subset=['average_N2O'], inplace=True)
```

#**Merging Dimensions and Tables**

**Merging Date**

```python
fact = pd.merge(covid_df, date_dim, how="left", on=["Date"] )
fact.drop(columns="Date", inplace=True)
```

**Merge and construct the Fact table**

```python
fact = pd.merge(fact, env_df, how="inner",on=["Year","Month"])
```

**Merging Country**

```python
fact = pd.merge(fact, country_data, how="inner",on="Country")
```

**Merging Government Response**

```python
fact = pd.merge(fact, gov_response_df, how="left",on=["Country"] )
```

**Merging World Economic Indicator**

```python
fact = pd.merge(fact, eco_data, how="left",on=["Country", "Year"])
```

```python
fact
```

#**Data Normalization/Scaling**

**Scaling Data World GDP to millions**

```python
fact.rename(columns={"World GDP": "World GDP(millions)"}, inplace=True)
fact['World GDP(millions)'] = fact['World GDP(millions)'].div(1000000)
```

**Scaling Area from square miles to square meters**

```python
fact.rename(columns={"Area (sq. mi.)": "Area(sq._km.)"}, inplace=True)
fact['Area(sq._km.)'] = fact['Area(sq._km.)'].mul(2.58999)
```

**Changing emissions so that changes are highlighted**

```python
fact.rename(columns={"average_CO2": "CO2 Change"}, inplace=True)
fact['CO2 Change'] = fact['CO2 Change'].sub(fact['CO2 Change'][0])

fact.rename(columns={"average_CO4": "CO4 Change"}, inplace=True)
fact['CO4 Change'] = fact['CO4 Change'].sub(fact['CO4 Change'][0])

fact.rename(columns={"average_N2O": "N2O Change"}, inplace=True)
fact['N2O Change'] = fact['N2O Change'].sub(fact['N2O Change'][0])
```

```python
fact
```

#**Feature Engineering**

**Adding percentage of population affected by covid**

```python
fact["covid_case_%"] = fact["total_cases"]*100/fact["Population"]
fact["covid_death_%"] = fact["total_deaths"]*100/fact["Population"]
fact["vaccinated__final_%"] = fact["num_vaccinated_final"]*100/fact["Population"]
fact["full_vaccinated__final_%"] = fact["num_fully_vaccinated_final"]*100/fact["Population"]
```

**Ensuring percentage caps at 100**

```python
fact.clip(upper=pd.Series({'covid_case_%': 100, "covid_death_%":100, "vaccinated__final_%":100, "full_vaccinated__final_%":100}), axis=1)
```

```python
fact
```

#**Data Transformation**

**Normalizing Emission Changes**

```python
fact["CO2 Change"] = (fact["CO2 Change"]-fact["CO2 Change"].min())/(fact["CO2 Change"].max()-fact["CO2 Change"].min())
fact["CO4 Change"] = (fact["CO4 Change"]-fact["CO4 Change"].min())/(fact["CO4 Change"].max()-fact["CO4 Change"].min())
fact["N2O Change"] = (fact["N2O Change"]-fact["N2O Change"].min())/(fact["N2O Change"].max()-fact["N2O Change"].min())
fact.rename(columns={"CO2 Change": "CO2 Change n", "CO4 Change":"CO4 Change n", "N2O Change":"N2O Change n"}, inplace=True)
```

**Normalizing Covid Percent Info**

```python
fact["covid_case_%"] = (fact["covid_case_%"]-fact["covid_case_%"].min())/(fact["covid_case_%"].max()-fact["covid_case_%"].min())
fact["covid_death_%"] = (fact["covid_death_%"]-fact["covid_death_%"].min())/(fact["covid_death_%"].max()-fact["covid_death_%"].min())
fact["vaccinated__final_%"] = (fact["vaccinated__final_%"]-fact["vaccinated__final_%"].min())/(fact["vaccinated__final_%"].max()-fact["vaccinated__final_%"].min())
fact["full_vaccinated__final_%"] = (fact["full_vaccinated__final_%"]-fact["full_vaccinated__final_%"].min())/(fact["full_vaccinated__final_%"].max()-fact["full_vaccinated__final_%"].min())
fact.rename(columns={"covid_case_%":"covid_case_% n", "covid_death_%":"covid_death_% n", "vaccinated__final_%":"vaccinated__final_% n", "full_vaccinated__final_%":"full_vaccinated__final_% n"}, inplace=True)
```

```python
fact
```

#**Final Steps**

**Make Columns Snake Case**

```python
fact.rename(columns={"Country":"Country_Name"}, inplace=True)
fact.columns = fact.columns.str.capitalize().str.replace(' ', '_')
```

**Replacing NaN**

```python
fact.fillna("N/A")
```

**Generating Surrogate Key**

```python
fact['id'] = pd.RangeIndex(start=1, stop=len(fact)+1)
fact = fact.reindex(columns=['id'] + list(fact.columns[:-1]))
```

**Exporting Dataset**

```python
fact.to_csv("Output.csv", index=False)
```
